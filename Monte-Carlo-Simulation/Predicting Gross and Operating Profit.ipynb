```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

# Revenue

```python
# Annual revenue of Ford from 2009 to 2021 in billion U.S. dollars (macrotrends.net)
rev_ford = [136.341, 127.144, 155.900, 160.338, 156.776, 151.800, 149.558, 144.077, 146.917, 133.559, 135.605, 128.954, 116.283]
```

```python
# Expected revenue
rev_mean = np.mean(rev_ford)
rev_mean
```

```python
# Expected standard deviation of revenue
rev_std = np.std(rev_ford)
rev_std
```

```python
# number of iterations 
iterations = 1000
```

```python
# Simulate 1000 future possible revenues
rev = np.random.normal(rev_mean, rev_std, iterations)
```

```python
plt.figure(figsize=(15, 6))
plt.plot(rev)
plt.title("Distribution of possible revenues", fontsize = 14, weight = "bold")
plt.show()
```

# Cost of Good Sold

```python
# Annual COGS of Ford from 2009 to 2021 in billion U.S. dollars (macrotrends.net)
cogs_ford = [114.651, 112.752, 134.693, 136.269,131.321, 126.195, 124.446, 125.025, 120.190, 113.039, 113.611, 104.451, 98.866]
```

```python
# Calculate COGS % of revenue
cogs_rev_ratio = []

for i in range(len(rev_ford)):
    cogs_rev_ratio.append(cogs_ford[i] / rev_ford[i])
```

```python
# Expected COGS ratio
cogs_ratio_mean = np.mean(cogs_rev_ratio)
cogs_ratio_mean
```

```python
# Expected standard deviation of COGS ratio
cogs_ratio_std = np.std(cogs_rev_ratio)
cogs_ratio_std
```

```python
# Simulate 1000 future possible COGS
cogs_ratio = np.random.normal(cogs_ratio_mean, cogs_ratio_std, iterations)
```

```python
COGS = - (rev * cogs_ratio)
```

```python
plt.figure(figsize=(15, 6))
plt.plot(COGS)
plt.title("Distribution of possible COGS", fontsize = 14, weight = "bold")
plt.show()
```

```python
COGS.mean()
```

```python
COGS.std()
```

# Gross Profit

```python
gross_profit = rev + COGS

plt.figure(figsize=(15, 6))
plt.plot(gross_profit)
plt.title("Distribution of possible gross profits", fontsize = 14, weight = "bold")
plt.show()
```

```python
max(gross_profit)
```

```python
min(gross_profit)
```

```python
gross_profit.mean()
```

```python
gross_profit.std()
```

```python
sns.set_style("white")
plt.figure(figsize = (10, 6)) 
plt.hist(gross_profit, bins = 20)
plt.title("Distribution of possible gross profit outcomes", fontsize = 14, weight = "bold")
plt.xlabel("Gross Profit in bn. $")
plt.ylabel("Number of Observations")
sns.despine()
plt.show()
```
